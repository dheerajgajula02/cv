```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread("photo.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread("test_2.jpg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img2=cv2.resize(img2, (70,100))
plt.imshow(img2)
```

```python
plt.imshow(img1)
```

```python
x_start = img1.shape[1]-img2.shape[1]
y_start = img1.shape[0]-img2.shape[0]
```

```python
img1.shape
```

```python
img2.shape
```

```python
#region of interest

roi = img1[y_start:img1.shape[0] , x_start:img1.shape[1]]
plt.imshow(roi)
```

```python
img2_gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
plt.imshow(img2_gray, cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2_gray)
plt.imshow(mask_inv, cmap='gray')
```

```python
mask_inv.shape
```

```python
import numpy as np
```

```python
black_back = np.full(img2.shape, 255, dtype= np.uint8)
```

```python
plt.imshow(black_back)
```

```python
black_back.shape
```

```python
bk = cv2.bitwise_or(black_back, black_back, mask= mask_inv)
```

```python
plt.imshow(bk)
```

```python
fg = cv2.bitwise_or(img2, img2, mask = mask_inv)
```

```python
plt.imshow(fg)
```

```python
final_roi = cv2.bitwise_or(roi, fg)
```

```python
plt.imshow(final_roi)
```

```python
final_roi.shape
```

```python
img1[y_start:img1.shape[0], x_start:img1.shape[1]]=final_roi
```

```python
plt.imshow(img1)
```
