```python
img1 = cv2.imread("../DATA/dog_backpack.png")
img1=cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread("../DATA/watermark_no_copy.png")
img2=cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2=cv2.resize(img2, (600,600))
```

```python
img1.shape
```

```python
x_offset = img1.shape[1]-img2.shape[1]
y_offset = img1.shape[0]-img2.shape[0]
```

```python
roi = img1[y_offset:img1.shape[0], x_offset:img1.shape[1]]
plt.imshow(roi)
```

when you create a gray scale image we are essentially taking the back version of that image but our aim is reverse the mask has to allow all the things except the backones though it so we're gonna complement the gray scale image

```python
img2_gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
plt.imshow(img2_gray, cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2_gray)
plt.imshow(mask_inv, cmap='gray')
```

```python
mask_inv.shape
```

as it no longer contains that color channels we have to rebuilt it

```python
import numpy as np
```

we basically apply bitwise_or (1+0 = 1) to all three layers so that they blend in together and produce an image that has colour channels and same image

```python
whiteback = np.full(img2.shape, 255, dtype=np.uint8)
```

```python
final_background = cv2.bitwise_or(whiteback, whiteback, mask =mask_inv)
plt.imshow(final_background)
```

```python
final_foreground = cv2.bitwise_or(img2,img2, mask=mask_inv)
plt.imshow(final_foreground)
```

```python
final_roi = cv2.bitwise_or(roi, final_foreground)
plt.imshow(final_roi)
```

```python
img1[y_offset:img1.shape[0], x_offset:img1.shape[1]]= final_roi
```

```python
plt.imshow(img1)
```
