```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../DATA/dog_backpack.jpg")
plt.imshow(img)
```

```python

drawing = False
ix,iy=-1,-1

def draw_rect(event , x,y, flags, params):
    global drawing, ix,iy, new_img
    
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing= True
        ix =x
        iy=y
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            img2 = cv2.imread("../DATA/dog_backpack.jpg")
            cv2.rectangle(img2, (ix,iy), (x,y), color=(0,0,255), thickness =5)
            new_img=img2
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        img2 = cv2.imread("../DATA/dog_backpack.jpg")
        cv2.rectangle(new_img, (ix,iy), (x,y), color=(0,0,255), thickness =5)
        new_img=img2


new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

cv2.namedWindow(winname='just_another')
cv2.setMouseCallback('just_another', draw_rect)



while True:
    cv2.imshow('just_another', new_img)
    
    if cv2.waitKey(20) & 0xFF==27:
        break
cv2.destroyAllWindows()




```

```python
image = cv2.imread("../DATA/dog_backpack.jpg")

drawing = False
ix,iy = -1,-1

def draw_rect(event, x,y, flags, params):
    global ix, iy, drawing, img
    
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix =x
        iy =y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            image_2 = cv2.imread("../DATA/dog_backpack.jpg")
            cv2.rectangle(image_2, pt1=(ix,iy), pt2=(x, y),color=(0,255,255),thickness=10)
            image = image_2
        
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        image_2 = cv2.imread("../DATA/dog_backpack.jpg")
        cv2.rectangle(image_2, pt1=(ix,iy), pt2=(x, y),color=(0,255,255),thickness=10)
        image = image_2
        
    
cv2.namedWindow(winname= "dog_image")
cv2.setMouseCallback("dog_image", draw_rect)

while True:
    cv2.imshow("dog_image", image)
    
    if cv2.waitKey(10) & 0xFF==27:
        break

cv2.destroyAllWindows()
    
```
