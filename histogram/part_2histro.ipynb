```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rainbow = cv2.imread("../DATA/rainbow.jpg")
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)
```

```python
rainbow.shape
```

```python
rainbow.shape[:2]
```

```python
mask=np.zeros(rainbow.shape[:2], np.uint8)
plt.imshow(mask, cmap='gray')
```

```python
mask[300:400, 100:400]=255
plt.imshow(mask, cmap='gray')
```

```python
plt.imshow(show_rainbow)
```

```python
masked_img = cv2.bitwise_and(rainbow, rainbow, mask=mask)
show_masked_img = cv2.bitwise_and(show_rainbow, show_rainbow, mask=mask)
```

```python
plt.imshow(show_masked_img)
```

```python
plt.imshow(cv2.cvtColor(masked_img, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(masked_img)
```

```python
cal_hist_red_mask = cv2.calcHist([rainbow], channels=[2],mask=mask, histSize=[256], ranges =[0,256] )
plt.plot(cal_hist_red_mask)
plt.title("masked red")
```

```python
cal_hist_red= cv2.calcHist([rainbow], channels=[2], mask=None, histSize=[256], ranges=[0,256] )
plt.plot(cal_hist_red)
plt.title("unmaked red")
```
