```python
img_1 = cv2.imread("../DATA/crossword.jpg", 0)
plt.imshow(img_1, cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img_1)
```

```python
img_1.min()
```

```python
img_1.max()
```

```python
import math
mi = round((img_1.min()+img_1.max())/2)
```

```python
mi
```

```python
th, bin_img_1 = cv2.threshold(img_1, 150, img_1.max(), cv2.THRESH_BINARY)
```

```python
show_pic(bin_img_1)
```

```python
thresh_2 = cv2.adaptiveThreshold(img_1, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 7,12)
show_pic(thresh_2)

#notice that here we are not using tuple unpacking
```

```python
fin_img = cv2.addWeighted(src1=bin_img_1, alpha = 0.6, src2= thresh_2, beta=0.4, gamma=0)
show_pic(fin_img)
```

```python
#adaptive thresholding
```

```python
thresh_2_adap = cv2.adaptiveThreshold(img_1, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 7,12)
```

```python
show_pic(thresh_2_adap)
```

```python
sec_blend_pic= cv2.addWeighted(bin_img_1, 0.6, thresh_2_adap, 0.4,0)
```

```python
show_pic(sec_blend_pic)
```
