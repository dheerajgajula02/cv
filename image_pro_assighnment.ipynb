```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
def display_img(img, cmap=None):
    fig = plt.figure(figsize=(10,12))
    ax=fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
girrafe= cv2.imread("DATA/giraffes.jpg")
show_girrfe = cv2.cvtColor(girrafe, cv2.COLOR_BGR2RGB)
display_img(show_girrfe)
```

```python
grey_girrafe=cv2.cvtColor(girrafe, cv2.COLOR_BGR2GRAY)

display_img( grey_girrafe, cmap='gray')
```

```python
grey_girrafe.max()
```

```python
grey_girrafe.min()
```

```python
ret, bin_girrafe= cv2.threshold(grey_girrafe, (grey_girrafe.max()+grey_girrafe.min())/2, grey_girrafe.max(),cv2.THRESH_BINARY)
```

```python
display_img(bin_girrafe, 'gray')
```

```python
ret
```

```python
hsv = cv2.cvtColor(girrafe, cv2.COLOR_BGR2HSV)
display_img(hsv)
```

```python
kernel = np.ones(shape=(4,4), dtype=np.float32)/10
```

```python
kernel
```

```python
blur_girrafe = cv2.filter2D(girrafe, -1, kernel)
```

```python
display_img(blur_girrafe)
```

```python
show_blur_girrafe = cv2.cvtColor(blur_girrafe, cv2.COLOR_BGR2RGB)
display_img(show_blur_girrafe)
```

```python
display_img(grey_girrafe, 'gray')
```

```python
sobel_x = cv2.Sobel(grey_girrafe, cv2.CV_64F, dx=1, dy=0, ksize=5)
display_img(sobel_x, cmap='gray')
```

```python
sobel_y = cv2.Sobel(grey_girrafe, cv2.CV_64F, dx=0, dy=1, ksize=5)
display_img(sobel_y, 'gray')
```

```python
blended_girrafe= cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, gamma=0)
display_img(blended_girrafe, 'gray')
```

```python
color=('b', 'g', 'r')

for i,col in enumerate(color):
    histr=cv2.calcHist([girrafe], [i], None,[256],[0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,256])
plt.title("girrrrrafeee")
```
